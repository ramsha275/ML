```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.metrics import mean_absolute_error as MAE
```

```python
df = pd.read_csv("train_bm.csv")
```

### Shuffling and Creating train test model

#### a) Shuffling:


```python
#Playing with the speed of execution (Pandas' SAMPLE vs Scikitlearn's SHUFFLE ) 

%time df.sample(frac=1 , random_state=1)
```

```python
%time sklearn.utils.shuffle(df , n_samples=df.shape[0])
```

```python
#We'll go with sklearn's shuffle method

new_df = sklearn.utils.shuffle(df ,random_state =1, n_samples=df.shape[0])
```

```python
new_df.head()
```

#### b) Spliting

```python
# Creating 4 divisions
div = df.shape[0]//4

# Splitting 3 parts in train and 1 part in test

train_df = new_df.iloc[:3*div+1,:]
test_df = new_df.iloc[3*div+1:,:]
```

```python
train_df.shape , test_df.shape , new_df.shape , ("final:",train_df.shape[0]+test_df.shape[0])
```

### Simple Mean ( mean of Item_Outlet_Sales)

```python
# storing simple mean in a new column in the test set as "simple_mean"
test_df['simple_mean'] = train_df['Item_Outlet_Sales'].mean()
```

```python
test_df.head()
```

```python
mae_item_out_sales = MAE(test_df['Item_Outlet_Sales'] , test_df['simple_mean'])
mae_item_out_sales
```

### Mean `Item Outlet Sales` with respect to `Outlet_Type`

```python
pd.pivot_table(train_df, values = 'Item_Outlet_Sales' , index = 'Outlet_Type' , aggfunc=np.mean)
```

```python
#Initializing new column 
test_df['Outlet_type_mean'] = 0

for i in train_df.Outlet_Type.unique():
    test_df['Outlet_type_mean'][test_df['Outlet_Type'] == str(i)] = train_df['Item_Outlet_Sales'][train_df['Outlet_Type'] == str(i)].mean()
```

```python
mae_out_type = MAE(test_df['Item_Outlet_Sales'] , test_df['Outlet_type_mean'])
mae_out_type
```

### Mean `Item Outlet Sales` with respect to `Outlet_Establishment_Year`

```python
pd.pivot_table(train_df, values = 'Item_Outlet_Sales' , index = 'Outlet_Establishment_Year' , aggfunc=np.mean)
```

```python
# initializing new column to zero
test_df['Outlet_year_mean'] = 0

for i in train_df.Outlet_Establishment_Year.unique():
    test_df['Outlet_year_mean'][test_df['Outlet_Establishment_Year'] == i] = train_df['Item_Outlet_Sales'][train_df['Outlet_Establishment_Year'] == i].mean()
```

```python
mae_out_year = MAE(test_df['Item_Outlet_Sales'] , test_df['Outlet_year_mean'])
mae_out_year
```

## Mean Item Outlet Sales with respect to Outlet_Location_Type

```python
pd.pivot_table(train_df, values = 'Item_Outlet_Sales', index = ['Outlet_Location_Type'], aggfunc = np.mean)
```

```python
# Initializing empty column
test_df['outlet_loc_mean'] = 0

# For every unique entry in Item_Identifier
for i in train_df['Outlet_Location_Type'].unique():
  # calculate and assign mean corresponding to the uniques entries
  test_df['outlet_loc_mean'][test_df['Outlet_Location_Type'] == str(i)] = train_df['Item_Outlet_Sales'][train_df['Outlet_Location_Type'] == str(i)].mean()
```

```python
#calculating mean absolute error
mae_out_loc = MAE(test_df['Item_Outlet_Sales'] , test_df['outlet_loc_mean'] )
mae_out_loc
```

## Mean Item_Outlet_Sales with respect to both Outlet_Location_Type and Outlet_Establishment_Year

```python
pd.pivot_table(train_df, values = 'Item_Outlet_Sales', index = ['Outlet_Location_Type','Outlet_Establishment_Year'], aggfunc = np.mean)
```

```python
# Initiating new empty column
test_df['Super_mean'] = 0

# Assigning variables to strings ( to shorten code length)
s2 = 'Outlet_Location_Type'
s1 = 'Outlet_Establishment_Year'

# For every Unique Value in s1
for i in test_df[s1].unique():
  # For every Unique Value in s2
  for j in test_df[s2].unique():
    # Calculate and Assign mean to new column, corresponding to both unique values of s1 and s2 simultaneously
    test_df['Super_mean'][(test_df[s1] == i) & (test_df[s2]==str(j))] = train_df['Item_Outlet_Sales'][(train_df[s1] == i) & (train_df[s2]==str(j))].mean()
```

```python
#calculating mean absolute error
super_mean_error = MAE(test_df['Item_Outlet_Sales'] , test_df['Super_mean'] )
super_mean_error
```
