```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/rianrajagede/iris-python/master/Datasets/iris/iris.csv")
df.head()
```

```python
df.shape
```

# Univariate Analysis :
In this we take 1 feature (here using sepal-length) and do analysis like is there any range/pattern between values and each class

```python
df['species'].unique()
```

```python
setosa = df[df.species == 'Iris-setosa']
viriginca = df[df.species == 'Iris-virginica']
versicolor = df[df.species == 'Iris-versicolor']
```

```python
# The 'o' will make dotss bigger
plt.plot(setosa['sepal_length'] , np.zeros_like(setosa['sepal_length']) ,'v', color='red')
plt.plot(viriginca['sepal_length'] , np.zeros_like(viriginca['sepal_length']) ,'+', color='green' )
plt.plot(versicolor['sepal_length'] , np.zeros_like(versicolor['sepal_length']) ,':' , color='y')

plt.xlabel("Sepal Length")
plt.ylabel("0-Scale")
plt.show()
```

### Conclusion  for Univariate:
- Setosa's sepal length is in between 4.3 - 5.8
- Virginca's sepal length is in between 4.9 - 7.9
- Versicolor's sepal length is in between 4.9 - 7.0
- All of them are overlapping 

# BiVariate Analysis:
In this we take 2 features (Petal_length , Speal_Length)


For FacetGrid -- https://seaborn.pydata.org/generated/seaborn.FacetGrid.html

```python
# FacetGrid can be used for multiple mini plots. It will just create a graph. It works on categorical 
# data only. If you pass rows and column it will create rxc graphs. 
# like setosa-versicolor, setosa-setosa , setosa, viriginca
# versicolor- setosa, versicolor-versicolor, versicolor-virginica
# virginica-setosa , virginica-versiocolor and virginica-virginica

sns.FacetGrid(df, col = 'species', row = 'species', hue = "species" , size=5)
```

```python
# if you want to draw a graph (inserting values) you have to use .map() and pass x and y axis
 
sns.FacetGrid(df, hue = "species" , size=5).map(plt.scatter,"petal_length","sepal_width").add_legend();
```

### Conclusion:
- With Setosa we can draw straight line. 
- But can't draw a line with Versicolor and Viriginca because they are over-lapping
